```python
import os
import json
import pandas as pd
```

```python
with open('config.json', 'r') as file:
    config = json.load(file)
```

```python
data = []
files = os.listdir(config["sources"])
for f in files:
    if not f.endswith('.csv'):
        continue
    tmp = pd.read_csv(config["sources"] + f, skiprows=6)
    tmp = tmp[["#hour_index", "hour_avg"]]
    tmp["hour_avg"] = tmp["hour_avg"].apply(lambda x: f"{str(x)}%")
    date = f.split("_")[3]
    tmp["_év_"] = date[:4]
    tmp["_hónap_"] = date[4:6]
    tmp["_nap_"] = date[6:8]
    # tmp["_dátum_"] = date[:8]
    tmp["_dátum_"] = tmp['#hour_index'].apply(lambda x: f"{date[:4]}.{date[4:6]}.{date[6:8]} {x} óra/h")
    data.append(tmp)
data = pd.concat(data)
data = data.reset_index(drop=True)
data = data.rename(columns = {"#hour_index" : "_óra_", "hour_avg" : "_átlag_"})
data.tail(24)
```

```python
data.to_json(f"{config['sources']}/total.json", orient="records")
```
